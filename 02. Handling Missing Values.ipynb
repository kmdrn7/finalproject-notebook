```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("/media/kmdr7/Seagate/DATASETS/IOT-23/Final Dataset/removed_string_dataset.csv")
dataset
```

```python
# Cek apakah ada missing values
dataset.isna().any().any()
```

## Replace Missing with Null/None

```python
# Replace missing values with Null
replace_null_dataset = dataset.replace(to_replace=np.nan, value=None)
replace_null_dataset.to_csv('/media/kmdr7/Seagate/DATASETS/IOT-23/Merged/02. Handling Missing Values/null_dataset.csv', index=False)
replace_null_dataset
```

## Replace Missing with Mean

```python
replace_mean_dataset = dataset.fillna(dataset.mean)
replace_mean_dataset.to_csv('/media/kmdr7/Seagate/DATASETS/IOT-23/Merged/02. Handling Missing Values/mean_dataset.csv', index=False)
replace_mean_dataset
```

## Replace Missing with Linear Regression

```python
replace_linear_dataset = dataset.interpolate(method ='linear', limit_direction ='forward')
replace_linear_dataset.to_csv('/media/kmdr7/Seagate/DATASETS/IOT-23/Merged/02. Handling Missing Values/linear_dataset.csv', index=False)
replace_linear_dataset
```

## Replace Missing with Drop

```python
replace_drop_dataset = dataset.dropna()
replace_drop_dataset.to_csv('/media/kmdr7/Seagate/DATASETS/IOT-23/Merged/02. Handling Missing Values/drop_dataset.csv', index=False)
replace_drop_dataset
```
