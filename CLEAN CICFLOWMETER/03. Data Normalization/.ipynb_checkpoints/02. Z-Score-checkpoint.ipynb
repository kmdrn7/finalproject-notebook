```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
def plotds(bef, aft):
    print("Sebelum")
    print("Min")
    plt.plot(bef.min())
    plt.show()
    print("Max")
    plt.plot(bef.max())
    plt.show()
    print("Sesudah")
    print("Min")
    plt.plot(aft.min())
    plt.show()
    print("Max")
    plt.plot(aft.max())
    plt.show()
```

# Z-Score Normalization

```python
scaler = StandardScaler()
```

```python
dataset = pd.read_csv("/media/kmdr7/Seagate/DATASETS/IOT-23-CIC/Dataset/2-Zero.csv")
label = dataset[["Label"]]
clean = dataset.drop(["Label"], axis=1)
minmax = pd.DataFrame(scaler.fit_transform(clean), columns=clean.columns)
plotds(clean, minmax)
minmax["Label"] = label
minmax.to_csv("/media/kmdr7/Seagate/DATASETS/IOT-23-CIC/Dataset/3-ZScore-2-Zero.csv", index=False)
```
