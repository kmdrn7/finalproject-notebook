```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
```

```python
dataset = pd.read_csv("/media/kmdr7/Seagate/DATASETS/IOT-23-CIC/Dataset/3-ZScore-2-Drop.csv")
clean_dataset = dataset.drop(columns=["Label"])
clean_dataset
```

# Principal Component Analysis (PCA)

```python
pca = PCA(0.9)
pca_dataset = pd.DataFrame(pca.fit_transform(clean_dataset))
pca_dataset["Label"] = dataset["Label"]
pca_dataset
```
